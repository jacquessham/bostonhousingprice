## Let's do some data cleaning

We have two columns of category variables:<br>
1) chas <br>
2) rad <br>
Since chas is a binary variable to indicate whether the houses are close to Charles River or not, we don't need to do further engineering.<br>
However, we may need to do feather engineering on rad. There are 9 classes in rad and we believe each class is independent of each other. One-hot-encoding would be suitable for this column.

```python
rad_dummies = pd.get_dummies(boston['rad'], prefix="rad")
rad_dummies.head()
```

Then merge back to boston dataframe.

```python
boston = pd.concat([boston, rad_dummies], axis=1).drop(columns=['rad'])
boston.head()
```

## Split data into train and test sets

```python
from sklearn.model_selection import train_test_split
```

First declare the features data frame and respond variable. Note that it is okay to have pandas dataframe and numpy array to fit into sklearn models.

```python
X = boston.drop(['medv'], axis=1)
y = boston['medv']
```

We will set the random_state to 49 for splitting the data and tree based model for the convenience of compare the models.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=49)
```

```python
X_train.head()
```

```python
y_train.head()
```

#### Model Evaluation
Since the our models are going to be regression based models, we will use RMSE to compare the models. The best model would have the lowest RMSE among the models.<br>
At the same time, we will also look at the score() to see the $R^2$ of the model

## First Model
#### Linear Regression with only one variable

Linear Regression start with only one feature to fit with the response variable. There could be 13 combination to fit for simple linear regression model (Linear Regression model with only one feature). In this experiment, we will only choose 1 the most important. Let's say the number of rooms.

```python
X_train_slr = pd.DataFrame(X_train['rm'])
X_test_slr = pd.DataFrame(X_test['rm'])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
slr = LinearRegression()
slr.fit(X_train_slr,y_train)
```

```python
slr.score(X_test_slr,y_test)
```

First Model explains 51% of the variation of the data.

```python
slr_pred = slr.predict(X_test_slr) # Obtain the fitted value
sqrt(mean_squared_error(y_test, slr_pred)) # Take the square root of MSE
```

The RMSE of First Model is 6.87

## Second Model
#### Linear Regression with all variable

Using only one feature for linear regression has very high variance when making prediction, we could include more feature in the model and thus produce a better model.

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.score(X_test, y_test)
```

First Model explains 71% of the variation of the data.

```python
lr_pred = lr.predict(X_test)
sqrt(mean_squared_error(y_test, lr_pred))
```

The RMSE of Second Model is 5.25.

#### Problem to Second Model

The problem to include all feature in Second Model that the Second Model may be overfitted. The solutions could be dropping feature(s) or regularization. We will explore more how to drop feature(s) in Second Model in the R code.

## Third Model
#### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor(random_state=49)
dt.fit(X_train,y_train)
```

```python
dt.score(X_test, y_test)
```

Third Model explains 83% of the variation of the data.

```python
dt_pred = dt.predict(X_test)
sqrt(mean_squared_error(y_test, dt_pred))
```

The RMSE of Third Model is 4.00

## Forth Model
#### Random Forest 

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(random_state=49)
rf.fit(X_train,y_train)
```

```python
rf.score(X_test, y_test)
```

Forth Model explains 88% of the variation of the data.

```python
rf_pred = rf.predict(X_test)
sqrt(mean_squared_error(y_test, rf_pred))
```

The RMSE of the Forth Model is 3.46

# Result
Based on the RMSE, we found the RMSE of random forest model has the lowest RMSE among all.
